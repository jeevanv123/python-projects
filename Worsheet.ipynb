```python
start_date=date(2022,4,1)
end_date=date(2022,7,1)
d=end_date-start_date

dates=[]
days=[]
for i in range(d.days):
  day=start_date+timedelta(days=i)
  dates.append(day)
  days.append(calendar.day_name[dates[i].weekday()])
```

```python
new_dates=[]
for i in dates:
  new_dates.append(date.isoformat(i))
```

```python
def hours():
  hh=pd.pivot_table(a,values=['Hours'],columns='Employee',aggfunc=np.sum,index='Month')
  return(hh)
```

```python
def pay():
  pp=pd.pivot_table(a,values=['Pay'],columns='Employee',aggfunc=np.sum,index='Month')
  return(pp)
```

```python
def employee(x):
  print("Employee : ",x)
  print("Employee ID: ",data[data['Employee_name']==x]['Employee_ID'].values[0])
  print("Position : ",data[data['Employee_name']==x]['Position'].values[0])
  print("Total days worked: ",a[a['Employee']==x].count()[1])
  return(pd.pivot_table(a[a['Employee']==x],values=['Pay','Hours'],aggfunc=np.sum,index='Month'))
```
